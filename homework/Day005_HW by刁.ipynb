```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import requests
import json

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')

datas=json.loads(r.text)

for data in datas:
    print("標題:",data['title'])
    print("貼文時間:",data['createdAt'])
    print("留言人數:",data['commentCount'])
    print("按讚人數:",data['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

#熱門文章

datas=json.loads(r.text)
length=len(datas)
totallike=0
totalcomment=0

for data in datas:
    totallike+=data['likeCount']
    totalcomment+=data['commentCount']

print('平均留言人數=', totalcomment/length ,'平均按讚人數=', totallike/length )
```

```python
#非熱門文章

r2=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
datas=json.loads(r2.text)
length=len(datas)
totallike=0
totalcomment=0

for data in datas:
    totallike+=data['likeCount']
    totalcomment+=data['commentCount']

print('平均留言人數=',totalcomment/length ,'平均按讚人數=',totallike/length)
```
