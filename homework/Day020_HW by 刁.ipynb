```python
soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = dict({})
for tr in table.find_all('tr'):
    if tr.find("td",attrs={'class':'no-alt'}):
        tag = tr.find("td",attrs={'class':'no-alt'}).text
        
        record=[]
        for td in tr.find_all("td"):
            record.append(td.text)
        d.setdefault(tag,{})
        d[tag][record[-3]] = record[-2]   #<td>2019/01</td><td>4.10</td><td> </td>   2019/01為record[-3] , 4.10為record[-2] , 空白為record[-1]
    else:
        if tr.find("td"):
            record = []
            for td in tr.find_all("td"):
                record.append(td.text)
            d[tag][record[0]] = record[1]
print(d)
```
