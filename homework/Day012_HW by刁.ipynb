### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url='https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)

soup=BeautifulSoup(r.text)

ds=soup.find(class_="part_list_2").find_all('h3')
    
for d in ds:
    print(d.find(class_="date").text , d.find(target="_blank" ).text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# dict
article = {}
# {類別: [文章標題]}

for d in ds:
    
    title=d.find(target="_blank").text
    tag=d.find('em').text
    
    if tag in article.keys(): # 已經出現過的文章類別
        article[tag].append(title)                           
    else:                         # 第一次出現的文章類別
        article[tag] = [title]                               # append是配合[]使用的  所以必須為[title]
               
article
```

### ③ 哪一個類別的文章最多

```python
max_count=0
max_tag=0


for key,value in article.items():               #article.items()  返回可遍历的(键, 值) 元组数组
    if len(value) > max_count:                  #判斷式:設一個變數，所有人跟他比，較大的就修改這個變數，然後就全部只比一次     
        max_count=len(value)
        max_tag = key
        
print(max_tag)    
```
