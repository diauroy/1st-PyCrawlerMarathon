## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from datetime import datetime, timedelta
import datetime
```

```python
date_time = datetime.datetime.now() - timedelta(1)  #昨天時間
date_time = date_time.strftime("%Y/%m/%d")

browser = webdriver.Chrome(executable_path='C:/Users/User/Desktop/chromedriver/chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

while True:
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 100000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)  #找當前網頁最後的時間，然後一值搜最後時間，到昨天0:00
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('STOP')
        break
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
        if date_time in d.find(class_="date").text :
            break
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
two_hours_ago_time = datetime.datetime.now() - timedelta(hours=2)
two_hours_ago_time = two_hours_ago_time.strftime("%Y/%m/%d")

browser = webdriver.Chrome(executable_path='C:/Users/User/Desktop/chromedriver/chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

while True:
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 100000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if two_hours_ago_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('STOP')
        break
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    
    if d.find(class_="date").text < two_hours_ago_time:
        print("stop")
        break
    print(d.find(class_="date").text , d.find('a').text ) 
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

three_days_ago_time = datetime.datetime.now() - timedelta(days=3)
three_days_ago_month = three_days_ago_time.strftime("%m")
three_days_ago_day = three_days_ago_time.strftime("%d")

#強制把時間改為三天前15點到17點
three_days_15_clock = datetime.datetime.strptime(three_days_ago_time.strftime("%Y/%m/%d") + " 15", '%Y/%m/%d %H')
three_days_17_clock = datetime.datetime.strptime(three_days_ago_time.strftime("%Y/%m/%d") + " 17", '%Y/%m/%d %H')
print(three_days_15_clock , three_days_17_clock)
```
